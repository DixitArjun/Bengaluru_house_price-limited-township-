```python
b=b.drop(['area_type','availability','society','balcony'],axis=1)
```

```python
b_filtered = b[(b['location'].isin(['Jayanagar','Koramangala','Electronic City'])) & (b['size'].isin(['3 BHK', '4 BHK']))]
```

```python
y=b_filtered.iloc[:,-1]
```

```python
b_filtered['size'] = b_filtered['size'].str[0]
```

```python
b_filtered['size']=b_filtered['size'].astype(int)
b_filtered['total_sqft']=b_filtered['total_sqft'].astype(int)
b_filtered['bath']=b_filtered['bath'].astype(int)
b_filtered['price']=b_filtered['price'].astype(int)
```

```python
b_filtered.info()
```

```python
b_filtered
```

```python
X = b_filtered.iloc[:, :-1].values
y = b_filtered.iloc[:, -1].values
```

```python
X
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
mr=LinearRegression()
mr.fit(X_train,y_train)
```

```python
y_pred=mr.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_pred),1)),1))
```
